```python
import pandas as pd
import numpy as np
import csv
import math
```

```python
def preprocess(file):
    
    
    raw=csv.reader(open(file))
    
    
    data=list(raw)
    
    return data
```

```python
raw=preprocess("yellow_tripdata_2015-08.csv")
```

```python
df=pd.DataFrame(raw)
```

```python
new_header = df.iloc[0]
df = df[1:] #take the data less the header row
df.columns = new_header
```

```python
df2=df.sample(800000)
```

```python
df2.to_csv(r'reduced yellow august.csv')
```
